```python
dbutils.widgets.dropdown("time_period", "weekly", ["weekly", "monthly", "daily"])
```

```python
dates = spark.sql("select min(date) as min_date, max(date) as max_date from sales").collect()[0]
today = dates["max_date"]      # <-- use dataset's latest date for calculations
```

```python
df_dates = spark.sql("select min(date) as min_date, max(date) as max_date from sales")
display(df_dates)
```

```python
from datetime import timedelta

time_period = dbutils.widgets.get("time_period")

if time_period == "weekly":
    start_date = today - timedelta(days=today.weekday())
    end_date = start_date + timedelta(days=6)

elif time_period == "monthly":
    start_date = today.replace(day=1)
    next_month = (today.replace(day=28) + timedelta(days=4)).replace(day=1)
    end_date = next_month - timedelta(days=1)

else:  # daily
    start_date = today
    end_date = today

print("Start:", start_date)
print("End:", end_date)
```

```python
display(spark.sql(
    f"""select city, sum(sale) 
        from sales 
        where date between '{start_date}' and '{end_date}' 
        group by city"""
))
```
